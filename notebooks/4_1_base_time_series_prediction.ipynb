```python
from statsmodels.tsa.arima.model import ARIMA
from itertools import product

combinations = product(range(3), range(3), range(3), range(2), range(2), range(2))
results = {}

for combination in combinations:
    p, d, q, P, D, Q = combination
    model = ARIMA(endog=train.wall_210, exog=train.bat_210, order=(p, d, q), seasonal_order=(P, D, Q, 12)).fit()
    results[model.aic] = combination

p, d, q, P, D, Q = results[min(results.keys())]
model = ARIMA(endog=train.wall_210, exog=train.bat_210, order=(p, d, q), seasonal_order=(P, D, Q, 12)).fit()

model.summary()
```

```python
from numpy import abs, mean

fc = model.forecast(test.shape[0], exog=test.bat_210, alpha=0.05)

mean(abs(fc - test.wall_210.to_numpy()) / abs(test.wall_210.to_numpy()) * 100)
```
